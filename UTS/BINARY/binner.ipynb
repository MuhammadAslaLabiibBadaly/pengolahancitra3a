### Library

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

### image open

```python
image = Image.open("asset/panda.jpg").convert("RGB")
image_np = np.array(image)
```

```python
print(image_np)
```

### Mengambil ukuran

```python
h, w, c = image_np.shape
```

```python
print(h)
print(w)
print(c)
```

### Inisialisasi Grayscale

```python
# Membuat array kosong untuk grayscale
grayscale_image = np.zeros((h, w), dtype=np.uint8)

# Konstanta untuk konversi ke grayscale
r_const, g_const, b_const = 0.299, 0.587, 0.114
```

### Melakukan grayscale menggunakan looping

```python
for i in range(h):
    for j in range(w):
        r, g, b = image_np[i, j]
        gray_value = int(r_const * r + g_const * g + b_const * b)
        grayscale_image[i, j] = gray_value
```

```python

print(grayscale_image)
```

### Menampilkan Output Grayscale

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(image_np)

plt.subplot(1,2,2)
plt.title("Grayscale Image")
plt.imshow(grayscale_image, cmap='gray')

plt.show()
```

### Inisialisasi binner

```python
# Membuat array kosong untuk biner
binary_image = np.zeros((h, w), dtype=np.uint8)

# Menentukan threshold (ambang batas)
threshold = 128  # Nilai tengah antara 0 - 255
```

### Melakukan looping binner dari grayscale

```python
for i in range(h):
    for j in range(w):
        if grayscale_image[i, j] >= threshold:
            binary_image[i, j] = 255
        else:
            binary_image[i, j] = 0 
```

```python
print(binary_image)
```

### Menampilkan Output Binner

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(image_np)
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Grayscale Image")
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Binary Image")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')

plt.show()
```
